# COVID19 Data Analysis

```python
#importing libraries

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

## Reading the data from csv 

```python
#loading the covid19_Confirmed_dataset
covid_data=pd.read_csv("covid19_Confirmed_dataset.csv")
covid_data.head()
```

```python
covid_data.shape
```

## Droping the useless columns from the data

```python
covid_data.drop(labels=['Lat','Long'],axis=1,inplace=True)
```

```python
covid_data
```

```python
covid_data.shape
```

## Aggregating the rows by country

```python
covid_data_aggregated=covid_data.groupby('Country/Region').sum()
covid_data_aggregated.head(10)
```

## visualizing data related to country

```python
covid_data_aggregated.loc['India'].plot()
```

```python
gp=covid_data_aggregated.loc[:,['4/21/20','4/22/20','4/23/20']]
print(gp.loc['India'])
gp.loc['India'].plot()
```

## Calculating and ploting first derivative of the curve

```python
covid_data_aggregated.loc['India'].diff().plot()
```

## Finding the maximum infected rate

```python
covid_data_aggregated.loc["India"].diff().max()
```

```python
countries=list(covid_data_aggregated.index)
max_infected_rates=[]
for country in countries:
    max_infected_rates.append(covid_data_aggregated.loc[country].diff().max())
covid_data_aggregated["max_infected_rates"]=max_infected_rates
covid_data_aggregated
```

```python
covid_data=pd.DataFrame(covid_data_aggregated['max_infected_rates'])
```

```python
covid_data.head()
```

# World wide happiness dataset

## Loding the world wide happiness dataset

```python
world_happiness_data=pd.read_csv("worldwide_happiness_report.csv")
world_happiness_data.head()
```

## drop useless columns in the dataset

```python
columns_to_drop=['Overall rank','Score','Generosity','Perceptions of corruption']
world_happiness_data.drop(columns_to_drop, axis=1, inplace=True)
world_happiness_data.head()
```

## changig th indices of the dataset

```python
world_happiness_data.groupby('Country or region')
world_happiness_data.head()
```

## join the two datasets that is prepared

```python
#covid data
covid_data.head()
```

```python
world_happiness_data=world_happiness_data.rename(columns={'Country or region':'Country/Region'})
```

```python
world_happiness_data.head()
```

```python
data=pd.merge(world_happiness_data,covid_data,on='Country/Region')
```

```python
data
```
